### Question Analysis

1. Which league has the highest revenue in the dataset?

```python
df_sort=df.sort_values (by='revenue',ascending=False)
print(df_sort)
```

IPL has the highest revenue in the dataset. the revenue is 7.997000e+09.

2. which league has the gross highest revenue?

```python
rev_high=df.groupby('leaguename')['revenue'].sum().sort_values(ascending=False)
rev_max=df.groupby('leaguename')['revenue'].sum().max()
print(rev_high,rev_max)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(rev_high)
plt.title('revenue of league')
plt.xlabel('league')
plt.ylabel('revenue')
plt.show()
```

The Premium league has the highest gross revenue.

3. which league has the lowest gross revenue?

```python
rev_low=df.groupby('leaguename')['revenue'].sum().sort_values(ascending=True)
rev_min=df.groupby('leaguename')['revenue'].sum().min()
print(rev_low,rev_min)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(rev_low)
plt.title('revenue of league')
plt.xlabel('league')
plt.ylabel('revenue')
plt.show()
```

NFL league has the lowest gross revenue.

14. How many viewers from india in each sports?

```python
plt.figure(figsize=(6,5))
plt.bar(india_viw.index, india_viw.values)
plt.title('viewers from india in each sports')
plt.xlabel('Sports')
plt.ylabel('number of viewers')
plt.xticks(rotation=90)
plt.show()
```

15. which league has the least number of teams? 

```python
team_low=df.groupby('leaguename')['totalteam'].sum().sort_values(ascending=True)
team_min=df.groupby('leaguename')['totalteam'].sum().min()
print(team_low,team_min)
```

```python
plt.figure(figsize=(6,5))
plt.bar(team_low.index, team_low.values)
plt.title('least total team by league')
plt.xlabel('League')
plt.ylabel('total teams')
plt.xticks(rotation=90)
plt.show()
```

IPL has the least least total team in the dataset

### Summary and Findings

This EDA project aimed to analyze a dataset containing information about various sports leagues, teams, and viewership. The project's objective was to gain insights into the data, identify patterns, and visualize the findings.

Findings
1. Top Sports Leagues: The analysis revealed that the top sports leagues in terms of total teams are the Premier League, NFL, Seria A, IPL, LA Liga etc

3. Team Distribution: The analysis revealed that the distribution of teams across leagues is uneven, with some leagues having a large number of teams while others have fewer teams.
4. Targeted Marketing: Leagues and teams can use the insights from this analysis to target their marketing efforts towards Indian viewers, who make up a significant proportion of the audience.
5. Investment in Indian Sports: The analysis highlights the potential for investment in Indian sports, particularly in cricket and football, which have a large and dedicated fan base.
